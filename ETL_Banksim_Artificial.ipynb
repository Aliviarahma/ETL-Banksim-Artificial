```python
import psycopg2
import psycopg2.extras as extras
import pandas as pd


def execute_values(conn, df, table):

    tuples = [tuple(x) for x in df.to_numpy()]

    cols = ','.join(list(df.columns))


    query = "INSERT INTO %s(%s) VALUES %%s" % (table, cols)
    cursor = conn.cursor()
    try:
        extras.execute_values(cursor, query, tuples)
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        cursor.close()
        return 1
    print("load success!")
    cursor.close()


# DB connection
conn = psycopg2.connect(
  database='postgres',
  user='postgres',
  password='password',
  host='localhost',
  port= '5432'

)
# Buat tabel di DB
sql = '''CREATE TABLE IF NOT EXISTS bsNET140513_032310(
Source varchar,Target varchar,Weight float,typeTrans varchar,fraud int) ;'''


cursor = conn.cursor()
cursor.execute(sql)
data = df2

data = data[['Source','Target','Weight','typeTrans','fraud']]

execute_values(conn, df2, 'bsNET140513_032310')
```

# Tranformation

## 1. Menggabungkan dua tabel dengan inner join
### Hal ini dilakukan karena kedua tabel memiliki shape yang sama dan saling terkait satu sama lain salah satunya dapat dilihat pada value kode custemer dan source

```python
df_main = pd.concat([df1, df2], axis=1, join='inner')
display(df_main)
```

```python
df_main.tail()
```

```python
df_main.isnull().sum()
```

## 2. Menghilangkan backtick pada value kolom
### Backtick dapat mengganggu format value yang sebenarnya sehingga dihilangkan agar ketika dilakukan explorasi value lebih mudah dibaca

```python
df_main['customer'] = df_main['customer'].replace("""[']""", '', regex=True)
```

```python
df_main.head()
```

```python
df_main['age'] = df_main['age'].replace("""[']""", '', regex=True)
df_main['gender'] = df_main['gender'].replace("""[']""", '', regex=True)
df_main['zipcodeori'] = df_main['zipcodeori'].replace("""[']""", '', regex=True)
```

```python
df_main['merchant'] = df_main['merchant'].replace("""[']""", '', regex=True)
df_main['zipmerchant'] = df_main['zipmerchant'].replace("""[']""", '', regex=True)
df_main['category'] = df_main['category'].replace("""[']""", '', regex=True)
df_main['source'] = df_main['source'].replace("""[']""", '', regex=True)
df_main['target'] = df_main['target'].replace("""[']""", '', regex=True)
df_main['typetrans'] = df_main['typetrans'].replace("""[']""", '', regex=True)
```

```python
df_main.head()
```

## 5. Mengganti kolom age kategori "U" diisi dengan value baru
### kategori "U" diketahui merupakan Unknown dalam hal ini kategori tersebut akan dihapus dan diganti dengan nilai null atau 0 kemudian dari nilai string di ubah ke integer, hal ini dilakukan agar bisa menghitung mean dari value tersebut. Lalu kategori "U" ini akan diisi dengan nilai mean dari age sehingga data tidak berkurang atau di hapus melainkan diisi dengan nilai baru yaitu rata - rata umur.

```python
#change "U" to 0
df_main2.drop(df_main2.loc[df_main2['age'] == "U"].index, inplace=True)
```

```python
df_main2.head()
```

```python
#convert to integer
df_main2['age'] = df_main2['age'].astype(int)
```

```python
avg_age = df_main2['age'].mean()
```

```python
#fill 0 value with average age
df_main2['age'] = df_main2['age'].fillna(avg_age)
```
